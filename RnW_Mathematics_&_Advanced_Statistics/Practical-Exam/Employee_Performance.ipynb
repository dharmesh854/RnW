#### 1️⃣ Define Mean, Median, Mode with an example in the context of employee salaries.

##### 📊 Mean (Average):
###### - The mean is calculated by adding all salary values and dividing by the total number of employees.

##### Formula:
#### Mean = ∑x / n

##### Example:
###### If five employees earn ₹50,000, ₹60,000, ₹70,000, ₹80,000, and ₹90,000:

###### 👉 Mean = (50,000 + 60,000 + 70,000 + 80,000 + 90,000) / 5 = ₹70,000

##### 📍 Median:
###### - The median is the middle value when salaries are arranged in ascending order.

##### Example:
###### ₹50,000, ₹60,000, ₹70,000, ₹80,000, ₹90,000

###### 👉 Median = ₹70,000

##### 🔁 Mode:
###### - The mode is the value that occurs most frequently.

##### Example:
###### ₹60,000, ₹60,000, ₹70,000, ₹80,000, ₹90,000

###### 👉 Mode = ₹60,000

#### 2️⃣ Differentiate between Range and Variance with examples.
##### 📏 Range
###### - Range measures the difference between the highest and lowest values.

##### Example:
###### Salaries = ₹50,000, ₹60,000, ₹70,000, ₹90,000

###### 👉 Range = 90,000 − 50,000 = ₹40,000

##### 📈 Variance
###### - Variance measures how much the values are spread around the mean.

##### Example:
###### If employee salaries vary greatly from the average salary, the variance is high. If salaries are close to the average, the variance is low.

### 🔍 Difference Between Range and Variance

##### | 📏 Range | 📈 Variance |
###### |-----------|------------|
###### | Uses only highest and lowest values | Uses all values |
###### | Simple measure of spread | Detailed measure of spread |
###### | More affected by extreme values | Gives overall variability |

#### 3️⃣ What is the difference between Normal Distribution and Poisson Distribution?
##### |🔔 Normal Distribution | 🎲 Poisson Distribution|
##### 
##### Continuous distribution | Discrete distribution
##### Bell-shaped and symmetric	| Usually right-skewed
##### Used for measurements | Used for counting events
##### Defined by mean and standard deviation | Defined by rate parameter (λ)
##### 💡 Example: 
###### 🔔 Normal Distribution: Employee Performance Scores in the dataset.
###### 🎲 Poisson Distribution: Number of customer complaints received per day.

#### 4️⃣ Explain Skewness and give one workplace-related example.
##### 📉 What is Skewness?

###### Skewness is a statistical measure  that shows the asymmetry or unevenness of a data distribution around it's mean.

##### Types:

###### ➡️ Positive Skew: Tail extends to the right.
###### ⬅️ Negative Skew: Tail extends to the left.
###### ⚖️ Zero Skew: Perfectly symmetric distribution.

##### 🏢 Workplace Example:

###### Most employees earn average salaries, while a few executives earn very high salaries.

###### 👉 This creates a positively skewed salary distribution.

#### 5️⃣ Define Conditional Probability and explain how it applies in promotions.
##### 🎯 Conditional Probability

###### - Conditional Probability is the probability of an event occurring given that another event has already occurred.

##### 🏢 Promotion Example:

###### Let:

###### A = Employee gets promoted 🎉
###### B = Employee has a Performance Score above 80 ⭐

##### Then:

##### 👉 P(Promotion | High Performance Score)

###### Represents the probability that an employee gets promoted given that they have a high performance score.

##### 📌 Application:

###### HR teams can use this to predict promotion chances based on employee performance.

#### 6️⃣ Explain Independent vs Mutually Exclusive Events with examples.
##### 🔹 Independent Events

###### - The occurence of one event doesn't affect the occurence of the other events.

##### Example:
###### An employee's department and another employee's project completion count are generally independent.

##### 🔹 Mutually Exclusive Events

###### - Two events are mutually exclusive if they cannot happen at the same time.

##### Example:
###### An employee's Promotion Status can be either:

###### ✅ Promoted
###### ❌ Not Promoted

###### Both cannot occur together.

##### 🔍 Difference
##### |🔹 Independent Events | 🚫 Mutually Exclusive Events|
###### Can occur together | Cannot occur together
###### One does not affect the other | One prevents the other

#### 7️⃣ What does Bayes Theorem help in real-world decision-making?
##### 🧠 Bayes Theorem
###### - Bayes’ Theorem is a way to update probabilities when you have new evidence.
######   It tells you how likely an event is given that another event has occurred.
###### or
###### - Bayes Theorem is a formula used to find the probability of an event with prior knowledge or related conditions.

##### 🏢 Real-World Example:

###### In HR analytics, Bayes Theorem can help estimate the probability of promotion based on:

###### ⭐ Performance Score
###### 📂 Projects Completed
###### ⏰ Working Hours

##### 🎯 Benefits:

###### ✅ Better predictions
###### ✅ Data-driven decisions
###### ✅ Continuous learning from new data

#### 8️⃣ Write a short note on PCA (Principal Component Analysis) in simple words.
##### 📊 What is PCA?

###### - PCA stands for 'Principal Component Analysis'. It's a dimensionality reduction technique that transforms the correlated variables in the smaller set of uncorrelated variables while preserving the variance.

##### 🏢 In Your Employee Performance Dataset:

##### Variables include:

###### 👤 Age
###### 💰 Salary
###### 📂 Projects Completed
###### ⏰ Working Hours
###### ⭐ Performance Score

###### PCA combines related variables into a smaller number of components, making analysis easier.

##### 🎯 Benefits of PCA:

###### ✅ Reduces data complexity
###### ✅ Removes redundant information
###### ✅ Speeds up analysis
###### ✅ Helps visualize high-dimensional data

##### 💡 Example:

###### Instead of analyzing 5 separate employee metrics, PCA may reduce them to 2–3 principal components while preserving most of the information.

#### Step 2 : Probability & Events
##### - Find the probability of employees getting promoted.
##### - Create a contingency table between Promotion_Status and Department.
##### - Compute conditional probability: (Promotion | Performance_Score > 80)

```python
# ---------------------------------------------------
# Find the probability of employees getting promoted
# ---------------------------------------------------

promoted_emp = df[df["Promotion_Status"] == "Yes"].shape[0]
total_emp = df["Employee_ID"].nunique()
print("Total Promoted Students: ",promoted_emp)
print("Total Employees: ",total_emp)

prob = promoted_emp/total_emp
print("Probability of Employees getting promoted: ",prob)
```

```python
# ------------------------------------------------------------------
# Create a contingency table between Promotion_Status and Department
# ------------------------------------------------------------------

contingency_table = pd.crosstab(df["Promotion_Status"], df["Department"])
contingency_table
```

```python
# ---------------------------------------------------------------------
# Compute conditional probability: (Promotion | Performance_Score > 80)
# ---------------------------------------------------------------------

promotion_yes = df[df["Promotion_Status"] == "Yes"]

score_above_80 = df[df["Performance_Score"] > 80]

yes_above_80 = df[
    (df["Promotion_Status"] == "Yes") & 
    (df["Performance_Score"] > 80)
    ]


p_promotion_given_above_80 = len(yes_above_80)/len(score_above_80)
print("Conditional Probability (Promotion | performance > 80): ",p_promotion_given_above_80)
```

```python
# ---------------------------------------
# Check Skewness and Kurtosis for Salary
# ---------------------------------------

# ----- Manual Skewness -----
n = len(emp_salary)
mean = np.mean(emp_salary)
median = np.median(emp_salary)
mode = emp_salary.mode()[0]
std = np.std(emp_salary)

third_moment = np.mean((emp_salary - mean)**3)
manual_skew = (third_moment/(std**3)) * (n**2 / ((n - 1) * (n - 2)))

print("Manual Skewness of Salary: ",manual_skew)


# ----- Scipy Skewness -----
scipy_skew = emp_salary.skew()
print("Scipy Skewness of Salary: ", scipy_skew)



# ----- Manual Kurtosis -----
fourth_moment = np.mean((emp_salary - mean)**4)
kurt_uncorrected = fourth_moment / (std**4)

bias_correction = ((n - 1) / ((n - 2) * (n - 3))) * ((n + 1) * (kurt_uncorrected * ((n-1)/n)) - 3 * (n - 1)) + 3
manual_kurt = bias_correction - 3  # Subtract 3 to make it "Excess" Kurtosis

print("Manual Kurtosis of Salary: ",manual_kurt)


# ----- Scipy Kurtosis -----
scipy_kurt = emp_salary.kurtosis()
print("Scipy Kurtosis of Salary: ",scipy_kurt)
```

#### Step 4 : Linear Algebra Application
##### Take the first 5 employees' [Projects_Completed, Working_Hours] as vector
##### - Perform dot product between two employee vectors.
##### - Find Norm 1 & Norm 2 of an employee's work vector.
##### - Compute the angle between two employees' work vector.

```python
# First 5 employees [Projects_Completed, Working_Hours] as vector
v1 = df[df["Employee_ID"] == "EMP100001"][["Projects_Completed","Working_Hours"]].values[0]
v2 = df[df["Employee_ID"] == "EMP100002"][["Projects_Completed","Working_Hours"]].values[0]
v3 = df[df["Employee_ID"] == "EMP100003"][["Projects_Completed","Working_Hours"]].values[0]
v4 = df[df["Employee_ID"] == "EMP100004"][["Projects_Completed","Working_Hours"]].values[0]
v5 = df[df["Employee_ID"] == "EMP100005"][["Projects_Completed","Working_Hours"]].values[0]

# -------------------------------------------------
# Perform dot product between two employee vectors
# -------------------------------------------------

# Manual Dot Product
dot_manual = v1[0]*v2[0] + v1[1]*v2[1]

# NumPy Dot Product
dot_numpy = np.dot(v1, v2)


# Output
print("Vector a:", v1)
print("Vector b:", v2)

print("\nManual Dot Product:", dot_manual)
print("NumPy Dot Product:", dot_numpy)
```

```python
# --------------------------------------------------
# Find Norm 1 & Norm 2 of an employee's work vector
# --------------------------------------------------


# ----------- Manual Calculation -----------
v3_L1_manual = np.sum(np.abs(v3)) 
v3_L2_manual = np.sqrt(np.sum(v3**2))

v4_L1_manual = np.sum(np.abs(v4)) 
v4_L2_manual = np.sqrt(np.sum(v4**2))
# ----------- Numpy Calculation -----------
v3_L1_numpy = np.linalg.norm(v3, ord=1)
v3_L2_numpy = np.linalg.norm(v3, ord=2)

v4_L1_numpy = np.linalg.norm(v4, ord=1)
v4_L2_numpy = np.linalg.norm(v4, ord=2)

print("Vector:", v3)
print("Vector:", v4)

print("\n----------------------------")
print("L1 & L2 Norms of Vector 3:")
print("----------------------------")

print("\nL1 Norm (Manual):", v3_L1_manual)
print("L1 Norm (NumPy):", v3_L1_numpy)

print("\nL2 Norm (Manual):", v3_L2_manual)
print("L2 Norm (NumPy):", v3_L2_numpy)

print("\n----------------------------")
print("L1 & L2 Norms of Vector 4:")
print("----------------------------")

print("\nL1 Norm (Manual):", v4_L1_manual)
print("L1 Norm (NumPy):", v4_L1_numpy)

print("\nL2 Norm (Manual):", v4_L2_manual)
print("L2 Norm (NumPy):", v4_L2_numpy)
```

```python
# ----------------------------------------------------
# Compute the angle between two employees' work vector
# ----------------------------------------------------

dot_product = np.dot(v4, v5)
mag_A = np.linalg.norm(v4)
mag_B = np.linalg.norm(v5)

cos_theta = dot_product / (mag_A * mag_B)
angle = np.degrees(np.arccos(cos_theta))

print("Angle between two vector (v4 & v5): ", angle)


plt.quiver(0, 0, v4[0], v4[1], angles="xy", scale_units="xy", scale=1, label="v4", color="blue")
plt.quiver(0, 0, v5[0], v5[1], angles="xy", scale_units="xy", scale=1, label="v5", color="orange")
plt.xlim(0, 50)
plt.ylim(0, 70)
plt.legend()
plt.grid(True, alpha=0.3)
plt.title("Angle Between Two Vectors")
plt.show()
```

##### 📊 Interpretation:
###### - The angle between vectors v4 and v5 is approximately 11.15°.
###### - Since the angle is very small, both vectors are pointing in nearly the same direction.
###### - This indicates a strong positive similarity/alignment between the two vectors.
