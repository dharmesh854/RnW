## Part A – Theoretical Foundation (Short Notes & Explaination)

### 1. What is Inferential Statistics ?
###### -> Inferential Statistics is the branch of statistics that allows us to make predictions or draw conclusions about a larger group by analyzing a smaller sample of data taken from it.

### 2. What is Hypothesis Testing and its components ?
###### -> Hypothesis Testing is a statistical method used to make decisions about population parameters based on sample data.

#### Components of Hypothesis Testing:
##### - Null Hypothesis (H0)
##### - Alternative Hypothesis (H1)
##### - Significance Level 
##### - Test Static

### 3. Explain Confidence Interval and Critical Value.
###### -> A Confidence Interval gives a range of values that is likely to contain the true population parameter based on sample data.
###### -> A critical value is the threshold used to decide whether to reject the null hypothesis.

### 4. Define P-Value.
###### -> the P-Value (Probability Value) is the probability of obtaining results at least  as extreme as the observed results, under the assumption that the null hypothesis (H0) is True.

### 5. Differentiate Type I and Type II errors.
###### -> Type I Error occurs when we reject the null hypothesis which as actually True.
###### -> Type II Error occurs when we fail to reject the null hypothesis which was actually False.

### 6. Brief Descriptions of z-test, t-test, chi-square test and ANOVA test.
###### - Z-test compares means when population variance is known or sample size is large.
###### - T-test compares means when population variance is unknown.
###### - Chi-Square test checks association between categorical variables.
###### - ANOVA compares means of three or more groups.

### 7. What is Covariance ?
###### -> Covariance measures how two variables change together.

### 8. What is Correlation ?
###### -> Correlation measures the strength and direction of the relationship between two variables.

## Part B - Data Analysis & Testing Tasks 

##### Import Libraries

```python
import pandas as pd
import numpy as np
from scipy import stats
```

##### Load Dataset

```python
df = pd.read_csv("Derivable_judgement_dataset.csv")

df.head()
```

##### 1. Formulate Hypotheses
##### Hypothesis 1

###### H₀: Smoking status has no effect on diabetes prevalence.

###### H₁: Smoking status affects diabetes prevalence.

##### Hypothesis 2

###### H₀: Mean BMI is equal across all age groups.

###### H₁: Mean BMI differs across age groups

##### 2. Confidence Interval for Age & Weight

```python
# ----- AGE -----
age_mean = df['age'].mean()
age_std = df['age'].std()
n = len(df)

ci = stats.t.interval(
    confidence=0.95,
    df=n-1,
    loc=age_mean,
    scale=age_std/np.sqrt(n)
)

print("95% Confidence Interval for Age:", ci)

# ----- WEIGHT -----
weight_mean = df['weight'].mean()
weight_std = df['weight'].std()

ci_weight = stats.t.interval(
    confidence=0.95,
    df=n-1,
    loc=weight_mean,
    scale=weight_std/np.sqrt(n)
)

print("95% Confidence Interval for Weight:", ci_weight)
```

##### 3. Find the Critical Value and p-value

##### T-Test Critical Value

```python
smokers = df[df['smoking_status']=="Smoker"]['bmi']
nonsmokers = df[df['smoking_status']=="Non-Smoker"]['bmi']

t_stat, p_value = stats.ttest_ind(smokers, nonsmokers)

dfree = len(smokers) + len(nonsmokers) - 2

critical_t = stats.t.ppf(0.975, dfree)

print("T Statistic =", t_stat)
print("Critical Value =", critical_t)
print("P Value =", p_value)
```

##### 4. Perform Z-Test or T-Test
##### T-Test (BMI of Smokers vs Non-Smokers)

```python
smokers = df[df['smoking_status']=="Smoker"]['bmi']
nonsmokers = df[df['smoking_status']=="Non-Smoker"]['bmi']

t_stat, p_value = stats.ttest_ind(smokers, nonsmokers)

print("T Statistic =", t_stat)
print("P Value =", p_value)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

##### 5. Conduct Chi-Square Test
##### Smoking Habit vs Diabetes

```python
table = pd.crosstab(
    df['smoking_status'],
    df['diabetes']
)

chi2, p_value, dof, expected = stats.chi2_contingency(table)

print("Chi Square Statistic =", chi2)
print("P Value =", p_value)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

##### 6. Perform ANOVA Test
##### BMI Across Age Groups

```python
groups = []

for name, group in df.groupby('age_group'):
    groups.append(group['bmi'])

f_stat, p_value = stats.f_oneway(*groups)

print("F Statistic =", f_stat)
print("P Value =", p_value)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

##### 7. Calculate Covariance and Correlation

```python
# ----- Covariance (Age vs BMI) -----
covariance = np.cov(
    df['age'],
    df['bmi']
)[0,1]

print("Covariance =", covariance)
```

```python
# ----- Correlation (Age vs BMI) ----- 
correlation = df['age'].corr(df['bmi'])

print("Correlation =", correlation)
```

##### 8. Clearly state the result and interpretation

```python
# ----- T-Test Interpretation -----
if p_value < 0.05:
    print("Reject H0: Smoking significantly affects BMI.")
else:
    print("Fail to Reject H0: Smoking does not significantly affect BMI.")
    
    
# ----- Chi-Square Interpretation -----
if p_value < 0.05:
    print("Reject H0: Smoking and Diabetes are associated.")
else:
    print("Fail to Reject H0: No significant association found.")
    
    
# ----- ANOVA Interpretation -----
if p_value < 0.05:
    print("Reject H0: BMI differs across age groups.")
else:
    print("Fail to Reject H0: BMI is similar across age groups.")
```
