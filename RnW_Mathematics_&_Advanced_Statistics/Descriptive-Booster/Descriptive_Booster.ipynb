## 3. Statistical Terms with Formula and Explanation

### A. Gaussian Distribution (Normal Distribution)

###### A Gaussian distribution is a bell-shaped symmetric distribution where most values are near the mean.

#### Characteristics:

###### - Mean = Median = Mode
###### - Symmetric curve
###### - Bell shaped

##### Example:
###### Household income may approximately follow normal distribution.

### B. Log Normal Distribution

###### A distribution where logarithm of values follows a normal distribution.

#### Characteristics:

###### - Right skewed
###### - Values cannot be negative

##### Example:
###### Income distributions are often log-normal.

### C. 3-Sigma Rule (Empirical Rule)

##### In a normal distribution:

###### 68% data lies within 1 standard deviation
###### 95% data lies within 2 standard deviations
###### 99.7% data lies within 3 standard deviations

##### Representation:

#### μ±1σ


#### μ±2σ


#### μ±3σ


### D. Percentiles

###### Percentiles divide data into 100 equal parts.

##### Example:

###### 90th percentile income means 90% households earn less than that value.

### E. Quartiles

###### Quartiles divide data into 4 equal parts.

##### Quartile	Meaning
###### Q1			25% of data
###### Q2			Median
###### Q3			75% of data

##### Example:

###### Q2 of income is median income.

### F. Five Number Summary

##### Five-number summary includes:

###### - Minimum
###### - Q1
###### - Median
###### - Q3
###### - Maximum

##### Purpose:

###### - Used in boxplots to understand data spread.

### G. Skewness

###### Skewness measures asymmetry of distribution.

##### Types:

###### - Positive skew → tail on right
###### - Negative skew → tail on left
###### - Zero skew → symmetric

### H. Kurtosis

###### Kurtosis measures peakedness of distribution.

##### Types:

###### - High kurtosis → sharp peak
###### - Low kurtosis → flat curve


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sci
```

```python
df = pd.read_csv("household_dataset_150.csv")
df
```

## Identify Numerical and Categorical Columns

```python
print("Numerical Columns:")
print(df.select_dtypes(include='number').columns)

print("\nCategorical Columns:")
print(df.select_dtypes(include='object').columns)
```

# Central Tendency

#### Central Tendency of 'Household_Income'

```python
income = df["Household_Income"]

print(f"Mean : {income.mean():,.0f}")        # :,.0f/:,.1f...:,.nf = Used to manage how many nums we want after point(.)
print(f"Median : {income.median():,.0f}")
print(f"Mode : {income.mode()[0]:,.0f}")
```

#### Central Tendency of 'Age_of_Household_Head'

```python
hh_age = df["Age_of_Household_Head"]

print(f"Mean : {hh_age.mean():,.1f} yrs")
print(f"Median : {hh_age.median()} yrs")
print(f"Mode : {hh_age.mode()[0]} yrs")
```

# Dispersion

#### Dispersion of 'Household_Income'

```python
print(f"Range : {income.max() - income.min()}")
print(f"Variance : {income.var():,.0f}")
print(f"Standard Deviation : {income.std():,.0f}")
```

#### IQR of 'Household_Income'

```python
Q1 = income.quantile(0.25)
Q3 = income.quantile(0.75)
IQR = Q3 - Q1
print()
print(f"Q1 (25th Percentile): {Q1}")
print(f"Q3 (75th Percentile): {Q3}")
print(f"IQR : {IQR}")
print()
print(f"The IQR of Household_Income is : {IQR}.")
```

# Histogram of 'Household_Income'

```python
# Histogram 
plt.hist(income, density = True , edgecolor='white', alpha=0.6, label='Income Distribution')  # alpha = Opacity 
    # (density=True so the curve overlays correctly)

# Fit and plot Gaussian Normal curve
mu, sigma = income.mean(), income.std()

x = np.linspace(income.min(), income.max(), 100)
y = sci.norm.pdf(x, mu, sigma)

plt.plot(x, y, color='black')

plt.title('Histogram of Household Income with Normal Curve')
plt.xlabel('Household Income')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```

# Skewness & Kurtosis

```python
skew = income.skew()
kurt = income.kurt()

print(f"Skewness : {skew:,.0f}")
print(f"Kurtosis : {kurt:,.0f}")

# Conclusion
sk_msg = "Right Skewed (Tail on right)" if skew > 0.5 else \
         "Left Skewed (Tail on left)" if skew < -0.5 else \
         "Approximately Symmetric"
         
krt_msg = "Heavy Tailed (Leptokurtic)" if kurt > 0 else \
          "Light Tailed (Platykurtic)" if kurt < 0 else \
          "Normal Tailed (Mesokurtic)"
          
print()
print(f"Skewness : {sk_msg}")
print(f"Kurtosis : {krt_msg}")
```

  # Boxplot by 'Education Level'

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# ── Boxplot: Income by Education Level ──
education_levels = ['Primary', 'Secondary', 'Graduate', 'Post-Grad']

income_data = [
    df[df['Education_Level'] == level]['Household_Income']
    for level in education_levels
]

axes[0].boxplot(income_data, labels=education_levels)

axes[0].set_title('Income by Education Level')
axes[0].set_xlabel('Education Level')
axes[0].set_ylabel('Household Income')

# ── Histogram: Urban vs Rural Income ──
urban_income = df[df['Urban_Rural'] == 'Urban']['Household_Income']
rural_income = df[df['Urban_Rural'] == 'Rural']['Household_Income']

axes[1].hist(urban_income, bins=12, alpha=0.6, label='Urban')
axes[1].hist(rural_income, bins=12, alpha=0.6, label='Rural')

axes[1].set_title('Urban vs Rural Income')
axes[1].set_xlabel('Household Income')
axes[1].set_ylabel('Frequency')
axes[1].legend()

# Show graphs
plt.tight_layout()
plt.show()

# ── Mean Income ──
print("Mean Income by Education Level:")
print(df.groupby('Education_Level')['Household_Income'].mean())

print("\nMean Income by Area:")
print(df.groupby('Urban_Rural')['Household_Income'].mean())
```
