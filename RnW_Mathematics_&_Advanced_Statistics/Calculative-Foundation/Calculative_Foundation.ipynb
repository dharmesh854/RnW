##### 2. Compute:
######   - Norm-1 and Norm-2 of vector
######   - Dot product and angle between two student's score vectors.
######   - Cross Product (for 3D selected subject)    

```python
# ----------------------------------------------------
#   Norm-1 and Norm-2 of vector 
# ----------------------------------------------------

std1f = std1.flatten()    # Convert 2D array into 1D vector

# ----------- Manual Calculation -----------
L1_manual = np.sum(np.abs(std1f)) 
L2_manual = np.sqrt(np.sum(std1f**2))

# ----------- Numpy Calculation -----------
L1_numpy = np.linalg.norm(std1f, ord=1)
L2_numpy = np.linalg.norm(std1f, ord=2)

print("Vector:", std1f)

print("\nL1 Norm (Manual):", L1_manual)
print("L1 Norm (NumPy):", L1_numpy)

print("\nL2 Norm (Manual):", L2_manual)
print("L2 Norm (NumPy):", L2_numpy)
```

```python
# ----------------------------------------------------
#   Dot Product and angle between two student's score vector 
# ----------------------------------------------------

std2_flat = std2.flatten()    # Student 2 vector
std3_flat = std3.flatten()    # Student 3 vector


# Manual Dot Product
dot_manual = std2_flat[0]*std3_flat[0] + std2_flat[1]*std3_flat[1]


# NumPy Dot Product
dot_numpy = np.dot(std2_flat, std3_flat)

# Magnitudes
mag_a = np.linalg.norm(std2_flat)
mag_b = np.linalg.norm(std3_flat)





# Angle Between Vectors
cos_theta = dot_numpy / (mag_a * mag_b)
theta = math.degrees(math.acos(cos_theta))
# angle = np.degrees(np.arccos(cos_theta))



print("Vector a:", std2_flat)
print("Vector b:", std3_flat)

print("\nManual Dot Product:", dot_manual)
print("NumPy Dot Product:", dot_numpy)

print("\nMagnitude of a:", mag_a)
print("Magnitude of b:", mag_b)

print("\nAngle between vectors (degrees):", theta)


# plt.quiver(0, 0, std2_flat[0], std3_flat[1], angles="xy", scale_units="xy", scale=1, color="blue", label="Student 2")
# plt.quiver(0, 0, std2_flat[0], std3_flat[1], angles="xy", scale_units="xy", scale=1, color="orange", label="Student 3")
# plt.xlim(0, 50)
# plt.ylim(0, 70)
# plt.title("Angle Between Two Vectors")
# plt.show()
```

##### 3. Find the projection of one vector onto another.

```python
# Students
std11 = np.array(df[df["Student_ID"] == "S11"][["Math","Physics","Chemistry","English","Computer_Science"]])
std12 = np.array(df[df["Student_ID"] == "S12"][["Math","Physics","Chemistry","English","Computer_Science"]])

# Student vectors
std11_flat = std11.flatten() 
std12_flat = std12.flatten() 


dot_product = np.dot(std11_flat, std12_flat)

# -----------------------------
# Step 2: Magnitude of b
# -----------------------------
magnitude_b = np.linalg.norm(std12_flat)

# -----------------------------
# Step 3: Scalar Projection
# -----------------------------
scalar_projection = dot_product / magnitude_b

# -----------------------------
# Step 4: Vector Projection
# -----------------------------
vector_projection = (dot_product / (magnitude_b**2)) * std12_flat

# -----------------------------
# Output
# -----------------------------
print("Vector a:", std11_flat)
print("Vector b:", std12_flat)

print("\nDot Product:", dot_product)
print("Scalar Projection of a onto b:", scalar_projection)
print("Vector Projection of a onto b:", vector_projection)


# Visualization
plt.quiver(0, 0, std11_flat[0], std11_flat[1], angles="xy", scale_units="xy", scale=1, color="blue")
plt.quiver(0, 0, std12_flat[0], std12_flat[1], angles="xy", scale_units="xy", scale=1, color="Orange")
plt.quiver(0, 0, vector_projection[0], vector_projection[1], angles="xy", scale_units="xy", scale=1)
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.title("Vector Projection")
plt.show()
```

##### Interpretation:

###### - Blue vector = Vector b
###### - Black vector = Vector a
###### - Orange segment = Projection of vector a onto vector b
###### - Dot Product = 30900, which is positive and very large.

##### What it means:

###### - The positive dot product indicates both vectors point in nearly the same direction.
###### - The projection measures how much of vector a lies along vector b.
###### - Since the projected vector is close to vector a, the two vectors are highly similar.

##### Conclusion:
###### 👉 The score patterns represented by vectors a and b are strongly aligned and highly correlated.

### Part D : Eigenvalues & Decomposition

##### 7. Compute the Eigenvectors and Eigenvalues of the covariance matrix.

```python
# -----------------------------
# Step 1: Eigenvalues & Eigenvectors
# -----------------------------
eigenvalues, eigenvectors = np.linalg.eig(matrix_A)

# -----------------------------
# Step 2: Verify Av = λv
# -----------------------------
v = eigenvectors[:,0]
lambda_val = eigenvalues[0]

Av = matrix_A @ v
lambda_v = lambda_val * v

# -----------------------------
# Output
# -----------------------------
print("Matrix A:\n", matrix_A)

print("\nEigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)

print("\nVerification (Av):", Av)
print("λv:", lambda_v)
```

##### 8. Perform LU Decomposition of the dataset matrix.

```python
# importing needed library for LU Decomposition
from scipy.linalg import lu

v07 = df[df["Student_ID"] == "S07"][["Math", "Physics","Chemistry"]].values[0]
v17 = df[df["Student_ID"] == "S17"][["Math", "Physics","Chemistry"]].values[0]
v13 = df[df["Student_ID"] == "S13"][["Math", "Physics","Chemistry"]].values[0]


matrix_C = np.array([v07, v17, v13])
print("Matrix C:")
print(matrix_C)
print()

P, L, U = lu(matrix_C)

# Verification
reconstructed_A = L @ U

print("Matrix A:\n", matrix_C)

print("\nLower Triangular Matrix L:\n", L)

print("\nUpper Triangular Matrix U:\n", U)

print("\nReconstructed A (L @ U):\n", reconstructed_A)
```

##### 9. Perform Singular Value Decomposition (SVD) and explain its role in dimensionality reduction.

```python
# -----------------------------
# SVD Decomposition
# -----------------------------
U, S, VT = np.linalg.svd(matrix_C)

# Convert singular values to diagonal matrix
Sigma = np.zeros_like(matrix_C, dtype=float)
np.fill_diagonal(Sigma, S)

# Reconstruct matrix
A_reconstructed = U @ Sigma @ VT

# Output
print("Matrix A:\n", matrix_C)

print("\nMatrix U:\n", U)

print("\nSingular Values:", S)

print("\nMatrix V^T:\n", VT)

print("\nReconstructed Matrix:\n", A_reconstructed)
```

### SVD's role in dimensionality reduction :
###### Singular Value Decomposition (SVD) helps reduce the number of dimensions in a dataset while preserving most of the important information.

##### Role:

###### - SVD decomposes a data matrix into components that capture the most significant patterns and variations in the data.
###### - It ranks these components by importance using singular values.
###### - Components with large singular values contain most of the information, while those with small singular values often represent noise or less important details.
###### - By keeping only the top few components and discarding the rest, SVD reduces the dataset's dimensions.
###### - This creates a smaller representation of the data that is easier to store, visualize, and process while retaining most of the original information.

##### 10. Apply Principal Component Analysis (PCA) to reduce the dataset from multiple subjects to 2 dimensions.

```python
# Importing needed libraries for PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

print("Matrix A:\n")
print(matrix_A)

# Standardize Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(matrix_A)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Output
print("Original Data:\n", matrix_A)

print("\nStandardized Data:\n", X_scaled)

print("\nPrincipal Components:\n", pca.components_)

print("\nExplained Variance Ratio:\n", pca.explained_variance_ratio_)

print("\nTransformed Data:\n", X_pca)
```

##### 11. Apply Linear Discriminant Analysis (LDA) to classify students into "Above Average" and "Below Average" categories.

```python
# Importing needed library for LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis


X = df[['Math',
        'Physics',
        'Chemistry',
        'English',
        'Computer_Science']]


y = df['Category'].map({
    'Below Average': 0,
    'Above Average': 1
})

# Apply LDA
# -----------------------------
lda = LinearDiscriminantAnalysis(n_components=1)

X_lda = lda.fit_transform(X, y)

# -----------------------------
# Output
# -----------------------------
print("Original Data:\n", X)

print("\nClass Labels:\n", y)

print("\nLDA Projection:\n", X_lda)

print("\nLDA Coefficients:\n", lda.coef_)
```
