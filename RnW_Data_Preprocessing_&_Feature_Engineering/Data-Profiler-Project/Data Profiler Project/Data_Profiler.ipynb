#### 4. Explain :
##### - What are tensors ?
##### - Provide an in-depth explanation of tensors with simple examples using NumPy ?

=> A Tensor is a mathematical object that generalizes scalars (0D), vectors (1D), and matrices (2D) to higher dimensions. In simple terms, a tensor is just a container for data in multiple dimensions, widely used in deep learning and numerical computing.

- Scalar (0D Tensor): A single number (e.g., 5).

- Vector (1D Tensor): A list of numbers (e.g., [1, 2, 3]).

- Matrix (2D Tensor): A grid of numbers with rows and columns.

- Tensor (3D+): Extends to more dimensions, such as a cube of numbers (3D) or higher.



=> 1. Scalar (0-D Tensor)

- A scalar contains only one value.

- Code: 
import numpy as np

scalar = np.array(10)

print(scalar)
print(scalar.ndim)

- Output:

10
0
- Explanation
Contains a single number, 
Dimension = 0

2. Vector (1-D Tensor)

- A vector is a collection of values in one dimension.

- Code: 
vector = np.array([10, 20, 30, 40])

print(vector)
print(vector.ndim)

- Output:

[10 20 30 40]
1

- Explanation
One row of values, 
Dimension = 1

3. Matrix (2-D Tensor)

- A matrix contains rows and columns.
- Code:
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(matrix)
print(matrix.ndim)

- Output:

[[1 2 3]
 [4 5 6]]
2

- Explanation
2 rows and 3 columns, 
Dimension = 2

4. 3-D Tensor

- A 3D tensor contains multiple matrices.

- Code:
tensor3d = np.array([
    [[1,2],[3,4]],
    [[5,6],[7,8]]
])

print(tensor3d)
print(tensor3d.ndim)

- Output:

[[[1 2]
  [3 4]]

 [[5 6]
  [7 8]]]
3

- Explanation

- Imagine:

Matrix 1:
1 2
3 4

Matrix 2:
5 6
7 8

- Two matrices stacked together form a 3D tensor.

## Part D : Exploratory Data Analysis (EDA)

#### Perform Univariate Analysis

- Numerical Variable

```python
print("\nAge Column:")
# Skewness
print("Skewness : ",df["Age"].skew())      # 0 = symmetric, >0 right, <0 left 
# Kurtosis
print("Kurtosis : ",df["Age"].kurtosis())
# IQR
print("IQR : ",df["Age"].quantile(0.75) - df["Age"].quantile(0.25))


print("\nIncome Column:")
# Skewness
print("Skewness : ",df["Income"].skew())      # 0 = symmetric, >0 right, <0 left 
# Kurtosis
print("Kurtosis : ",df["Income"].kurtosis())
# IQR
print("IQR : ",df["Income"].quantile(0.75) - df["Income"].quantile(0.25))


print("\nPurchases Column:")
# Skewness
print("Skewness : ",df["Purchases"].skew())      # 0 = symmetric, >0 right, <0 left 
# Kurtosis
print("Kurtosis : ",df["Purchases"].kurtosis())
# IQR
print("IQR : ",df["Purchases"].quantile(0.75) - df["Purchases"].quantile(0.25))
```

- Visualizing Numeric - Histogram

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# ----- Purchases -----

# Histogram
axes[0][0].hist(df["Purchases"], bins=30, color="#ed3a799f", edgecolor="black")
axes[0][0].set_title("Histogram (Purchases)")
axes[0][0].set_xlabel("Purchases")
axes[0][0].set_ylabel("Frequency")

# KDE Plot
sns.kdeplot(df["Purchases"], ax=axes[0][1], fill=True, color="#7c3aed", alpha=0.4)
axes[0][1].set_title("KDE Plot (Purchases)")

# Box Plot
axes[0][2].boxplot(df["Purchases"], patch_artist=True, boxprops=dict(facecolor="#cffafe", color="#0891b2"),
    medianprops=dict(color="#db2777", linewidth=2)
)
axes[0][2].set_title("Box Plot (Purchases)")

# ----- Income -----

# Histogram
axes[1][0].hist(df["Income"], bins=30, color="#ed3a799f", edgecolor="black")
axes[1][0].set_title("Histogram (Income)")
axes[1][0].set_xlabel("Income")
axes[1][0].set_ylabel("Frequency")

# KDE Plot
sns.kdeplot(df["Income"], ax=axes[1][1], fill=True, color="#7c3aed", alpha=0.4)
axes[1][1].set_title("KDE Plot (Income)")

# Box Plot
axes[1][2].boxplot(df["Income"], patch_artist=True, boxprops=dict(facecolor="#cffafe", color="#0891b2"),
    medianprops=dict(color="#db2777", linewidth=2)
)
axes[1][2].set_title("Box Plot (Income)")

# ----- Age -----

# Histogram
axes[2][0].hist(df["Age"], bins=30, color="#ed3a799f", edgecolor="black")
axes[2][0].set_title("Histogram (Age)")
axes[2][0].set_xlabel("Age")
axes[2][0].set_ylabel("Frequency")

# KDE Plot
sns.kdeplot(df["Age"], ax=axes[2][1], fill=True, color="#7c3aed", alpha=0.4)
axes[2][1].set_title("KDE Plot (Age)")

# Box Plot
axes[2][2].boxplot(df["Age"].dropna(), patch_artist=True, boxprops=dict(facecolor="#cffafe", color="#0891b2"),
    medianprops=dict(color="#db2777", linewidth=2)
)
axes[2][2].set_title("Box Plot (Age)")

plt.tight_layout()
plt.show()
```

#### 👤 Age Distribution (Histogram, KDE & Box Plot)
##### 🔍 Interpretation
- Most customers belong to a specific age group, indicating the dominant customer segment.
- The KDE plot shows the density of customers across different ages.
- The box plot helps identify age variability and potential outliers.
- Overall, the customer base is concentrated within a particular age range.

#### 💰 Income Distribution (Histogram, KDE & Box Plot)
##### 🔍 Interpretation
- Customer income is distributed across multiple income levels.
- Most customers fall within the middle-income category.
- The box plot highlights the spread of income and any extreme income values.
- Income variation indicates differences in customers' purchasing power.

#### 🛒 Purchases Distribution (Histogram, KDE & Box Plot)
##### 🔍 Interpretation
- The majority of customers make a moderate number of purchases.
- The KDE curve shows where purchase activity is most concentrated.

```python
# ─── Value Counts ───
print(df["gender"].value_counts())
print(df["gender"].value_counts(normalize=True) * 100)  # %

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ─── Bar Chart ───
category_counts = df["gender"].value_counts()
axes[0].bar(category_counts.index, category_counts.values,)
axes[0].set_title("Customers by Gender")
axes[0].set_xlabel("Gender"); axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

# ─── Seaborn Countplot (even easier!) ───
sns.countplot(data=df, x="gender",
              order=df['gender'].value_counts().index,
              ax=axes[1])
axes[1].set_title('Seaborn Countplot — Same Data, Less Code!')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

#### Perform Bivariate Analysis

- Categorical vs Numerical - Box & VIolin

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Relationship between Gender vs Purchases
df.boxplot(column="Purchases", by="gender", ax=axes[0][0],patch_artist=True,
            boxprops=dict(facecolor="#cffafe", color="#0891b2"),
            medianprops=dict(color="#db2777", linewidth=2)
)
axes[0][0].set_title("Purchases by Gender")
axes[0][0].set_xlabel("Gender")
axes[0][0].set_ylabel("Purchases")

sns.violinplot(x="gender", y="Purchases", data=df, palette="Set2", ax=axes[0][1])
axes[0][1].set_title("Purchases Distribution by Gender")
axes[0][1].tick_params(axis="x", rotation=45)



# Relationship between Churn vs Income
df.boxplot( column="Income", by="Churn", ax=axes[1][0], patch_artist=True,
            boxprops=dict(facecolor="#b4f8b8", color="#0891b2"),
            medianprops=dict(color="#db2777", linewidth=2)
)
axes[1][0].set_title("Income by Churn")
axes[1][0].set_xlabel("Churn")
axes[1][0].set_ylabel("Income")

sns.violinplot(x="Churn", y="Income", data=df, palette="Set2", ax=axes[1][1])
axes[1][1].set_title("Income Distribution by Churn")
axes[1][1].tick_params(axis="x", rotation=45)


plt.tight_layout()
plt.show()
```

#### 📊 Interpretation of the Visual

##### 🛍️ 1. Purchases by Gender (Boxplot & Violin Plot)
- 👨 Male and 👩 Female customers show very similar purchase behavior.
- The median number of purchases is nearly the same for both genders.
- The spread and distribution of purchases are also comparable, indicating that gender has little influence on purchase frequency.
- No significant differences or unusual variations are observed between the two groups.

#### 💰 2. Income by Churn (Boxplot & Violin Plot)
- 👥 Customers who churned ("Yes") and those who did not churn ("No") have similar income distributions.
- Median income levels are nearly identical across both groups.
- Income variability and overall spread are also very similar.
- This suggests that income is not a strong differentiating factor for customer churn.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ─── Scatter Plot ───
axes[0].scatter(
    df["Income"],
    df["Purchases"],
    alpha=0.5,
    color="#2563eb",
    edgecolors="white",
    linewidth=0.5
)

axes[0].set_xlabel("Income")
axes[0].set_ylabel("Purchases")
axes[0].set_title("Income vs Purchases")

# ─── Seaborn Regplot (Scatter + Regression Line) ───
sns.regplot(
    data=df,
    x="Income",
    y="Purchases",
    ax=axes[1],
    color="#7c3aed",
    scatter_kws={"alpha": 0.4},
    line_kws={"color": "#db2777", "linewidth": 2}
)

axes[1].set_title("Income vs Purchases — With Trend Line")

plt.tight_layout()
plt.show()

# ─── Correlation Value ───
corr = df["Income"].corr(df["Purchases"])
print(f"Pearson Correlation: {corr:.3f}")
```

#### 💰 Income vs Purchases (Scatter Plot)
##### 🔍 Interpretation
- The scatter plot shows the relationship between income and purchasing behavior.
- Customers with higher income generally tend to make more purchases.
- The distribution of points indicates the strength of the relationship.

#### Perform Multivariate Analysis

```python
plt.figure(figsize=(8, 6)) 
corr = df.select_dtypes(include="number").corr() 

# Hide upper triangle 
sns.heatmap( corr, annot=True, fmt='.2f', center=0, vmin=-1, cmap="RdYlGn", vmax=1, linewidths=0.5, square=True ) 
plt.title("Correlation Heatmap") 
plt.tight_layout()
plt.show()
```

#### 🌡️ Correlation Heatmap
##### 🔍 Interpretation
- The heatmap visualizes correlations among all numerical variables.
- Darker colors indicate stronger relationships.
- Positive values show variables moving in the same direction.
- Negative values indicate inverse relationships.
- It helps identify the most influential features in the dataset.

```python
plt.figure(figsize=(15, 10))

scatter = plt.scatter(df["Income"], df["Purchases"],
    c=df["Age"],          # Color represents Age
    s=df["tenure"] * 10,  # Size represents Tenure
    alpha=0.6, cmap="viridis",
    edgecolors="white", linewidth=0.5)

plt.colorbar(scatter, label="Age")

plt.xlabel("Income")
plt.ylabel("Purchases")
plt.title("Income vs Purchases (Age & Tenure Included)")

plt.show()
```
