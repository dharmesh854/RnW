```python
# Summary statistics for numerical columns
summary = df[['CreditScore','EstimatedSalary','Balance']].describe()
summary
```

```python
# Correlation Matrix
correlation_matrix = df[['CreditScore','EstimatedSalary','Balance']].corr()
correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Heatmap of correlarion matrix
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='vlag')
plt.title('HEAT MAP')
plt.show()
```

```python
# Top 5 transaction IDs by transaction
top_transaction = df.groupby('Surname')['EstimatedSalary'].sum().nlargest(5)
top_transaction
```

```python
# Plot the bar chart
top_transaction.plot(kind='bar', color=['pink','purple'], width = 0.4,alpha=0.6)
plt.title('Top 5  Customer IDs by their salary')
plt.xlabel('Customer Surname')
plt.ylabel('Estimated Salary')
plt.show()
```

```python
# Daily Transaction trend

line_chart = df.groupby('Surname')['CreditScore'].sum().nlargest(5)

line_chart
```

```python
# Plot the line chart
plt.figure(figsize=(10, 6))
line_chart.plot(kind='line', marker='o', markersize = 8,color = 'black', markerfacecolor = 'yellow')
plt.title('Top 5 Surnames by their Credit Score')
plt.xlabel('Surname')
plt.ylabel('Credit Score')
```

```python
# Sales by region
country = df.groupby('Geography')['CreditScore'].sum()

country
```

```python
# Plot the bar chart
country.plot(kind='bar', color=['blue','red','yellow'], alpha=0.6)
plt.title('Credit Score in Each Country')
plt.xlabel('Country')
plt.ylabel('Credit Score')
plt.show()

# Identify region with highest sales growth
growth_region = country.idxmax()
growth_value = country.max()
print(f'The region with the highest Credit score is {growth_region} with a CreditScore of {growth_value}.')
```
